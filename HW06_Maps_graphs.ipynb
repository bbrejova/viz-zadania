## Importing district boundaries

* Below we import GeoPandas table of Slovak districts (okresy) into variable `districts`.
* This data is under CC-BY 4.0 license by GKÚ Bratislava, it was originally downloaded from https://www.geoportal.sk/sk/zbgis_smd/na-stiahnutie/
* Each row is one district. Important columns for us are:
  * `IDN3`: unique numeric ID of a district (e.g. 602)
  * `NM3`: name of the district (e.g. Banská Štiavnica)
  * `NM2`: name of the region (kraj) to which this district belongs (e.g. Banskobystrický)
  * `geometry`: GPS coordinates of the district boundaries

```python
url = "https://fmfi-compbio.github.io/viz/data/districts.json"
districts = geopandas.read_file(url)
display(Markdown("**Dimensions of the `districts` table:**"))
display(districts.shape)
display(Markdown("**The first two rows of `districts` shown as columns:**"))
display(districts.head(2).transpose())
```
