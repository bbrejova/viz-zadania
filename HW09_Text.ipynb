## Finding the most common words using nltk library

Wordcloud library autonomously discovers words in the provided text, computes their frequencies, omits stop words and creates the figure. Now we will explicitly count and print word frequencies using nltk library to see the data underlying our clouds. You will need `common_word_set` variable created below in tasks 3 and 4.


````python
# again start by concatenating all overviews separated by spaces
text = " ".join(list(movies['overview']))
# convert text to lower case and split into words using nltk
tokens = nltk.word_tokenize(text.lower())
# keep only words consisting of letters (isalpha)
# and omit stop words
words = [w for w in tokens if w.isalpha() and not w in stopwords]

# find 10 most frequent words using nltk
# the result is a list of pairs (word, its count)
common_words = nltk.FreqDist(words).most_common(10)
# get the most frequent words as a set of strings
common_word_set = {word for (word, count) in common_words}

# display the results
display(Markdown("**List common_words** (words and their counts):"))
display(common_words)
display(Markdown("**Set common_word_set:**"))
display(common_word_set)
````
