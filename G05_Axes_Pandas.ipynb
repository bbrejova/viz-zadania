When using a logarithmic axis, we compute the logarithm of an original value and use it as a coordinate. Zero and negative numbers cannot be displayed because their logarithm is undefined.

### Selecting movie genre ratings

Columns from 20 to 30 of `fsev_data` contain preferences of respondents with respect to various movie genres, such as `Horror`, `Comedy` or `Action`. Each answer is a rating from 1 to 5, where 5 means the highest preference.

The code below selects these columns from the table to a new table `movies` and prints the mean rating for each genre sorted from the least to most popular. It also creates a long version of the table called `movies_long` with columns `person`, `genre`, `rating`. Column `person` contains the row numbers from the original table so that we know which values belong to the same person.

```python
movies = fsev_data.iloc[:, 20:31]
movies_long = (movies.reset_index()
              .rename({'index':'person'}, axis=1)
              .melt(id_vars="person", var_name="genre", value_name="rating"))
display(Markdown("**Mean popularity of movie genres:**"))
display(movies.mean().transpose().sort_values())
display(Markdown("**The first and last 3 rows of the movies_long table:**"))
display(movies_long.head(3))
display(movies_long.tail(3))
```
